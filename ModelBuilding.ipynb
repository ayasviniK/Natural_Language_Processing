# Load Data

```python
from sklearn import datasets
beast_cancer = datasets.load_breast_cancer()
print(beast_cancer)
```

```python
#assign data and target (class label)
x = beast_cancer.data
y = beast_cancer.target

#print the first 10 rows of data
print(x[:10])
```

# Data Preprocessing

```python
import pandas as pd

# Convert x to a DataFrame for easier missing value handling
data = pd.DataFrame(x, columns=beast_cancer.feature_names)

# Check for missing values
print(data.isnull().sum())

# Fill or drop missing values as necessary
data.fillna(method='ffill', inplace=True)
```

# Visualizing Data

```python
%pip install seaborn

import matplotlib.pyplot as plt
import seaborn as sns

# Visualize the distribution of the target variable using y
sns.countplot(x=y)
plt.title('Target Variable Distribution')
plt.xlabel('Target')
plt.ylabel('Count')
plt.show()
```

```python
# Scatter plot of two features
plt.scatter(x[:, 0], x[:, 1], c=y, cmap='coolwarm')
plt.xlabel(beast_cancer.feature_names[0])
plt.ylabel(beast_cancer.feature_names[1])
plt.title('Feature Scatter Plot')
plt.show()
```

# Splitting the Data

```python
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.2, random_state=1)
print(x_train.shape)
print(x_test.shape)
print(y_train.shape)
print(y_test.shape)
```

# Build the Model

```python
from sklearn import tree
classifier = tree.DecisionTreeClassifier(random_state=1)
```

# Train the model

```python
clf = classifier.fit(x_train, y_train)
```

# Display the model

```python
tree.plot_tree(clf, feature_names=beast_cancer.feature_names, class_names=beast_cancer.target_names, filled=True, rounded=True, proportion=True)
plt.show()
```

# Prediction & Evaluate the Model

```python
prediction = classifier.predict(x_test)
print("Prediction for test data")
print(prediction)
print("Actual values for test data")
print(y_test)
```

# Calculation accuracy

```python
from sklearn.metrics import accuracy_score
print("Accuracy is")
print(accuracy_score(y_test, prediction))
```

# Confusion Matrix

```python
from sklearn.metrics import confusion_matrix
print("Confusion Matrix")
print(confusion_matrix(y_test, prediction))
```

# Precision, Recall and F1 Score

```python
from sklearn.metrics import classification_report
target_names = beast_cancer.target_names
print(classification_report(y_test, prediction, target_names=target_names))
```

# Final Model (trained on all data)

```python
clf = classifier.fit(x, y)
```

```python
print("Single Prediction")
row = [x[0]]  # Example: first row
yhat = clf.predict(row)
print(f'Predicted Class: {yhat[0]}')
```
